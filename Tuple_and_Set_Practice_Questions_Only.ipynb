
### Question 8
Create a tuple with repeated values and count how many times an element appears.

```python
a= ("apple","banana" , "cherry" , "apple", "cherry")
s = []
for i in a:
  if i not in s:
    print(f"elements of {i} is:",a.count(i))
    s.append(i)
```
